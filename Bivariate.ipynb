```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
calc_vif(dataset[["etest_p","salary"]])
```

```python
# `degree_p` ஓட VIF value 112 ங்குறது, அது மத்த independent variables கூட ரொம்ப strongly correlate ஆகி இருக்குனு காட்டுது. 
# அதாவது, அது தனிப்பட்ட information ஐ விட, மத்த variables ஓட information ஐயே திரும்ப கொடுக்குது.

# அதனால, `degree_p` ஐ remove பண்றதுனால, modelக்கு redundant information குறையும்.
 #   இது model ஓட stability & interpretability ஐ improve பண்ணும்.
```
